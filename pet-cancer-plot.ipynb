## Graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create the DataFrame with the values from the website plot
years = [2024, 2025, 2026, 2027, 2028, 2029]
values = [0.39, 0.43]

# Extend the values to match the years with annual compuond growth 10.8 % growth till 2028
values.extend([0.43 * (1 + 0.108) ** i for i in range(1, 3 + 1)])

# Add the last value to match the years
values.append(0.65)

# Create a DataFrame
df = pd.DataFrame({'Year': years, 'Value': values})
df
```

```python
# Plot the data
fig,  ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(df['Year'], df['Value'], color='#386D61') # Extracted the color from the website plot

# Annotate the bars with their values, the website plot only has the first two values and the last one (but I included all)
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval +
                0.01, f'{yval:.2f}', ha='center', va='bottom')

# Add title and labels
ax.set_title(
    'Pet Cancer Therapeutics Global Market Report 2025', fontsize=14)
ax.set_ylabel('Market Size (in USD billion)')

ax.set_ylim(0, max(values) + 0.1)  # Start y-axis at 0 (very important for bar charts)

ax.set_axisbelow(True)  # Draw grid lines below the bars
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

plt.savefig('pet_cancer_therapeutics_market_report_2025.png', dpi=300, bbox_inches='tight')
plt.show()
```
